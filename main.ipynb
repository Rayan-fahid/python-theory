## esercizio 5


```python
import random

# Creiamo una lista vuota per contenere i numeri
numeri_casuali = []

# Generiamo 10 numeri interi casuali (ad esempio tra 1 e 50) e li salviamo nella lista
for i in range(10):
    numero = random.randint(1, 50)
    numeri_casuali.append(numero)

# Stampiamo la lista originale per vedere i numeri generati
print("Lista originale generata casualmente:")
print(numeri_casuali)
print("-" * 40)

# Scorriamo la lista usando gli indici perché dobbiamo MODIFICARE i valori interni
for i in range(len(numeri_casuali)):
    # Verifichiamo se il numero nella posizione 'i' è dispari
    if numeri_casuali[i] % 2 != 0:
        # Se è dispari, raddoppiamo il suo valore direttamente nella lista
        numeri_casuali[i] = numeri_casuali[i] * 2

# Stampiamo la lista modificata
print("Lista con i soli valori dispari raddoppiati:")
print(numeri_casuali)
```
